```python
import pandas as pd
from sklearn.model_selection import train_test_split
import spacy

# Preprocess text
nlp = spacy.load('en_core_web_lg')
def preprocess_text(text):
    doc = nlp(text)
    # Tokenization, Lowercasing, and Lemmatization
    processed_tokens = [token.lemma_.lower() for token in doc]
    return processed_tokens

df["phrase1_processed"] = df["phrase1"].apply(preprocess_text)
df["phrase2_processed"] = df["phrase2"].apply(preprocess_text)

df
```

```python
# Split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(df[["phrase1_processed", "phrase2_processed"]], df["label"], test_size=0.2)
```

```python
import numpy as np
```

```python
def get_phrase_embedding(text):
    # Join tokenized text into a single string
    joined_text = " ".join(text)
    doc = nlp(joined_text)  # Process joined text with spaCy
    word_embeddings = [token.vector for token in doc]
    phrase_embedding = np.mean(word_embeddings, axis=0)
    return phrase_embedding

df["phrase1_embedding"] = df["phrase1_processed"].apply(get_phrase_embedding)
df["phrase2_embedding"] = df["phrase2_processed"].apply(get_phrase_embedding)
```

```python
df
```

```python
# Choose your preferred metric
def get_similarity(embedding1, embedding2):
    # Cosine similarity
    return np.dot(embedding1, embedding2) / (np.linalg.norm(embedding1) * np.linalg.norm(embedding2))

df["similarity"] = df.apply(lambda row: get_similarity(row["phrase1_embedding"], row["phrase2_embedding"]), axis=1)
```

```python
df
```

```python
def is_similar(similarity_score, threshold):
    return 1 if similarity_score > threshold else 0

df["is_similar"] = df["similarity"].apply(lambda score: is_similar(score, 0.5))
```

```python
df
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Calculate evaluation metrics
accuracy = accuracy_score(df["label"], df["is_similar"])
precision = precision_score(df["label"], df["is_similar"])
recall = recall_score(df["label"], df["is_similar"])
f1 = f1_score(df["label"], df["is_similar"])

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1-score:", f1)
```
