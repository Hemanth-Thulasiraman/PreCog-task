```python
import gensim
```

```python
model = gensim.models.Word2Vec(workers=3)
```

```python
word2vec_model = model
def get_word_vector(word, model):
    try:
        return model.wv[word]
    except KeyError:
        # Handle out-of-vocabulary words
        return None

# Create new columns for word vectors
df['word1_vector'] = df['word1'].apply(lambda x: get_word_vector(x, word2vec_model))
df['word2_vector'] = df['word2'].apply(lambda x: get_word_vector(x, word2vec_model))
```

```python
df
```

```python
sentences = df[["word1", "word2"]].apply(lambda x: " ".join(x), axis=1).tolist()
```

```python
print(sentences)
```

```python
tokenized_sentences = [simple_preprocess(x) for x in sentences]
print(tokenized_sentences)
```

```python
import gensim
```

```python
model = gensim.models.Word2Vec(
    tokenized_sentences,
    vector_size=300,
    window=5,
    min_count=1,
    sg=1
)
```

```python
similarity = model.wv.similarity("old","new")
print(similarity)
```

```python
df["similarity_score"] = df[["word1", "word2"]].apply(
    lambda row: model.wv.similarity(row["word1"], row["word2"]), axis=1
)
```

```python
df
```

```python
df["scaled_similarity"] = (df["similarity_score"] + 1) / 2 * 9 + 1
```

```python
df
```

```python
from scipy.stats import spearmanr
```

```python
correlation, p_value = spearmanr(df["SimLex999"], df["scaled_similarity"])
print("SRCC:", correlation)
```
