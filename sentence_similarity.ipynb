```python
import pyarrow.parquet as pq
import pandas as pd

# Read the Parquet file using the correct path
table = pq.read_table("/content/train-00000-of-00001.parquet")

# Convert to Pandas DataFrame
df = table.to_pandas()

# Now you can work with the DataFrame 'df'
```

```python
import spacy
from scipy import spatial
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Load spaCy model
nlp = spacy.load("en_core_web_lg")  # Adjust model size as needed

# Preprocess text using spaCy
def preprocess_text(text):
    doc = nlp(text)
    tokens = [token.text for token in doc]
    return tokens

df["sentence1_processed"] = df["sentence1"].apply(preprocess_text)
df["sentence2_processed"] = df["sentence2"].apply(preprocess_text)
```

```python
df
```

```python
import numpy as np
```

```python
def get_sentence_embedding(text):
    # Join tokenized text into a single string
    joined_text = " ".join(text)
    doc = nlp(joined_text)  # Process joined text with spaCy
    word_embeddings = [token.vector for token in doc]
    sentence_embedding = np.mean(word_embeddings, axis=0)
    return sentence_embedding

df["sentence1_embedding"] = df["sentence1_processed"].apply(get_sentence_embedding)
df["sentence2_embedding"] = df["sentence2_processed"].apply(get_sentence_embedding)
```

```python
df
```

```python
# Choose your preferred metric
def get_similarity(embedding1, embedding2):
    # Cosine similarity
    return np.dot(embedding1, embedding2) / (np.linalg.norm(embedding1) * np.linalg.norm(embedding2))

df["similarity"] = df.apply(lambda row: get_similarity(row["sentence1_embedding"], row["sentence2_embedding"]), axis=1)
```

```python
def is_similar(similarity_score, threshold):
    return 1 if similarity_score > threshold else 0

df["is_similar"] = df["similarity"].apply(lambda score: is_similar(score, 0.5))
```

```python
df
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Calculate evaluation metrics
accuracy = accuracy_score(df["label"], df["is_similar"])
precision = precision_score(df["label"], df["is_similar"])
recall = recall_score(df["label"], df["is_similar"])
f1 = f1_score(df["label"], df["is_similar"])

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1-score:", f1)
```
